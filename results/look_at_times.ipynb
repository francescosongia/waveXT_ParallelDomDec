```python
import numpy as np
import matplotlib.pyplot as plt

def plot_fill(arr_list,title_list,title=''):
    n = len(arr_list)
    x = np.arange(1,n+1)
    means = np.zeros([1,n])
    stds = np.zeros([1,n])
    
    for i in range(n):
        means[0,i] = np.mean(arr_list[i])
        stds[0,i] = np.std(arr_list[i])

    plt.figure()  
    plt.plot(x,means[0,:],'.-')
    plt.fill_between(x, means[0,:] - stds[0,:], means[0,:] + stds[0,:], color='C0', alpha=0.5)
    plt.xticks(x, title_list)
    plt.title(title)
    plt.show()
```
